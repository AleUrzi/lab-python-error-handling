````python
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Quantity cannot be negative.")
                valid_quantity = True
                inventory[product] = quantity
            except ValueError as error:
                print(f"Error: {error}. Please enter a valid quantity.")
    return inventory

def get_customer_orders(products, inventory):
    valid_num_orders = False
    while not valid_num_orders:
        try:
            num_orders = int(input("Enter the number of products you want to order: "))
            if num_orders < 0:
                raise ValueError("Number of orders cannot be negative.")
            valid_num_orders = True
        except ValueError as error:
            print(f"Error: {error}. Please enter a valid number of orders.")

    customer_orders = set()
    for i in range(num_orders):
        valid_product = False
        while not valid_product:
            product = input(f"Enter the name of a product that a customer wants to order ({i+1}/{num_orders}): ").lower().strip()
            if product in products and inventory.get(product, 0) > 0:
                customer_orders.add(product)
                valid_product = True
            else:
                print(f"Error: '{product}' is not in the inventory or out of stock. Please enter a valid product name.")
    
    return customer_orders

def calculate_total_price(customer_orders):
    total_price = 0.0
    for product in customer_orders:
        valid_price = False
        while not valid_price:
            try:
                price = float(input(f"Enter the price of {product}: "))
                if price < 0:
                    raise ValueError("Price cannot be negative.")
                total_price += price
                valid_price = True
            except ValueError as error:
                print(f"Error: {error}. Please enter a valid price.")
    return total_price

def update_inventory(inventory, customer_orders):
    updated_inventory = {
        product: max(0, quantity - 1)  # Decrement quantity by 1 for each ordered product
        for product, quantity in inventory.items()
    }
    updated_inventory = {product: quantity for product, quantity in updated_inventory.items() if quantity > 0}
    return updated_inventory

# Main Program
products = ["t-shirt", "mug", "hat", "book", "keychain"]
inventory = initialize_inventory(products)
customer_orders = get_customer_orders(products, inventory)

# Display Order Statistics
total_products_ordered = len(customer_orders)
percentage_unique_ordered = (total_products_ordered / len(products)) * 100
print("\nOrder Statistics:")
print(f"Total Products Ordered: {total_products_ordered}")
print(f"Percentage of Unique Products Ordered: {percentage_unique_ordered:.1f}%")

# Update Inventory
inventory = update_inventory(inventory, customer_orders)

# Display Updated Inventory
print("\nUpdated Inventory:")
for product, quantity in inventory.items():
    print(f"{product}: {quantity}")

# Calculate and Display Total Price
total_price = calculate_total_price(customer_orders)
print(f"Total Price: {total_price:.2f}")
````
